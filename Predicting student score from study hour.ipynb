```python
import pandas as pd
import matplotlib.pyplot as plt  
from sklearn.linear_model import LinearRegression  
import numpy as np
from sklearn.metrics import mean_squared_error   
%matplotlib inline
```

```python
dataframe = pd.read_csv('student_scores.csv')
```

```python
dataframe.head(2)
```

```python
plt.plot(dataframe.Hours, dataframe.Scores)
```

```python
# for better visualization
plt.plot(dataframe.Hours, dataframe.Scores, 'o')
```

# Model building

```python
#split data to training and testing set, (we will do this manually)
X = dataframe.Hours.values.reshape(-1,1)
Y = dataframe.Scores.values.reshape(-1,1)
X_train, Y_train = X[0:20], Y[0:20]
X_test, Y_test = X[19:], Y[19:]
```

```python
print(X_train, Y_train)
```

```python
print(X_test, Y_test)
```

```python
#initializing our model
model = LinearRegression()
```

```python
#fitting our model
model.fit(X_train, Y_train)
```

```python
# plot the regression line
regression_line = model.predict(X)
plt.plot(X,regression_line, color = 'r')
plt.scatter(X,Y)
```

```python
# to see how the fit line relates to our training and testing data sets
regression_line = model.predict(X)
plt.plot(X,regression_line, color = 'r')
plt.scatter(X_train,Y_train)
plt.scatter(X_test,Y_test)
```

```python
y_predictions = model.predict(X_test)
```

```python
y_predictions
```

```python
# Lets us estimate the error
error = mean_squared_error(Y_test, y_predictions)
```

```python
error
```

```python
print(f'The MSE is: {error}')
```
